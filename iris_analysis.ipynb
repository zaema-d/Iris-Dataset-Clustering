```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# Determine the optimum number of clusters using the elbow method
wcss = []  # Within-Cluster Sum of Square
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(iris_df)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')  # Within cluster sum of squares
plt.show()
```

```python
def KmeanWithOptimalClusters(opt):
  # From the plot, choose the number of clusters where the elbow occurs
  optimal_clusters = opt

  # Apply K-means clustering with the optimal number of clusters
  kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
  y_kmeans = kmeans.fit_predict(iris_df)
```

```python
def plotCluster():
  # Visualize the clusters
  plt.figure(figsize=(10, 6))
  plt.scatter(iris_df.iloc[y_kmeans == 0, 0], iris_df.iloc[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
  plt.scatter(iris_df.iloc[y_kmeans == 1, 0], iris_df.iloc[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
  plt.scatter(iris_df.iloc[y_kmeans == 2, 0], iris_df.iloc[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')

  # Plot the centroids of the clusters
  plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroids')
  plt.title('Clusters of Iris Data')
  plt.xlabel(iris.feature_names[0])
  plt.ylabel(iris.feature_names[1])
  plt.legend()
  plt.show()
```

```python
#Clusters representation with the number of optimal clusters = 3
KmeanWithOptimalClusters(3)
plotCluster()
```

```python
#Clusters representation with the number of optimal clusters = 4
KmeanWithOptimalClusters(4)
plotCluster()
```

```python
#Clusters representation with the number of optimal clusters = 6
KmeanWithOptimalClusters(6)
plotCluster()
```

```python
#Clusters representation with the number of optimal clusters = 2
KmeanWithOptimalClusters(2)
plotCluster()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Set up the subplots
fig, axs = plt.subplots(1, 4, figsize=(15, 5))

# Plot for 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)
y_kmeans = kmeans.fit_predict(X)
axs[0].scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
axs[0].set_title('2 Clusters')

# Plot for 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)
axs[1].scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
axs[1].set_title('3 Clusters')

# Plot for 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)
axs[2].scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
axs[2].set_title('4 Clusters')

# Plot for 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
y_kmeans = kmeans.fit_predict(X)
axs[3].scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
axs[3].set_title('5 Clusters')



# Show the plots
plt.show()
```
